SGD

```python
def calc_mse(y, y_pred):
    err = np.mean((y - y_pred)**2)
    return err

def eval_model(X, y, iterations, alpha=1e-4):
    np.random.seed(42)
    W = np.random.randn(X.shape[0])
    n = X.shape[1]
    for i in range(1, iterations + 1):
        y_pred = np.dot(W, X)
        err = calc_mse(y, y_pred)
        W -= alpha * (1/n * 2 * np.dot((y_pred - y), X.T))
        if i % (iterations / 10) == 0:
            print(i, W, err)
    return W
```

```python
eval_model(X_st, y, iterations=1000, alpha=1e-3)
```

```python
def eval_sgd_model(X, y, iterations, qty_in_batch, alpha=1e-4):
    np.random.seed(42)
    W = np.random.randn(X.shape[0])
    n = X.shape[1] # число наблюдений
    n_batch = n // qty_in_batch
    if n % qty_in_batch != 0:
        n_batch += 1
    # print(n_batch)
    for i in range(1, iterations + 1):
        for b in range(n_batch):
            start_ = qty_in_batch * b
            end_ = qty_in_batch * (b + 1)
            X_tmp = X[:, start_ : end_]
            y_tmp = y[start_ : end_]
            y_pred_tmp = np.dot(W, X_tmp)
            err = calc_mse(y_tmp, y_pred_tmp)
            W -= alpha * (1/n * 2 * np.dot((y_pred_tmp - y_tmp), X_tmp.T))
        if i % (iterations / 10) == 0:
            print(i, W, err)
    return W
```

```python
eval_sgd_model(X_st, y, iterations=1000, qty_in_batch=7, alpha=1e-3) # 100.15781666892349
```

**1. Напишите функцию наподобие gradient_descent_reg_l2, но для применения L1-регуляризации.**

$$Q(w, X) + \lambda |w| \rightarrow \underset{w}{\text{min}}.$$


Производная = $$\frac{x}{|x|},$$ при $x \neq 0$ 

Также можно записать как $$sgn(x),$$ при $x \neq 0$ 

```python
def sgn(x):
    if x > 0:
        return 1 
    elif x < 0:
        return 0
    else:
        raise ValueError('Х не должен быть = 0')

def sgn_array(array):
    return np.array([sgn(_) for _ in array])
```

```python
def eval_model_reg1(X, y, iterations, alpha=1e-4, lambda_=1e-8):
    np.random.seed(42)
    W = np.random.randn(X.shape[0])
    n = X.shape[1]
    for i in range(1, iterations + 1):
        y_pred = np.dot(W, X)
        err = calc_mse(y, y_pred)
        W -= alpha * (1/n * 2 * np.dot((y_pred - y), X.T) + lambda_ * sgn_array(W)) 
        
    return W
```

```python
lambdas = np.linspace(0,30,50)

weights = [eval_model_reg1(X_st, y, iterations=1000, alpha=1e-3, lambda_=lambda_) for lambda_ in lambdas]
weights = np.array(weights)
```

```python
x_labels = ['int', 'стаж', 'стоимость занятия', 'количество предметов']

plt.ylabel('weight')
plt.xlabel('lambda')
plt.title('Lasso')

for w, label in zip(weights.T, x_labels):
    plt.plot(lambdas,w, label=label)

plt.legend()
plt.show()
```
